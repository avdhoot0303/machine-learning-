```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y= x ** 2
```

```python
x
```

```python
y
```

**functional method**

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')


plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

**OBJECT ORIENTED METHOD**

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('TITLE')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger plot")

axes2.plot(y,x)
axes2.set_title("Smaller plot")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')

plt.tight_layout()
```

**FIgure size and DPI**

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(4,3))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
axa = fig.add_axes([0,0,1,1])
axa.plot(x,x**2,label='X squared')
axa.plot(x,x**3,label='X cubed')

axa.legend(loc=0)
```

**PLOT APPEARANCE**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='black',linewidth=3,alpha=0.5,linestyle='-',marker='o',
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') #also accepts RGB HEXCODES
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='black',linewidth=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
