**Seaborn regression Plots**

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

**Linear model plot**

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex')
```
